```python
movies_dic = {}
n = int(input("Enter the Number of Users :"))

for i in range(n):
    movies = ()
    name = str(input("Enter the Name of User: "))
    movies = str(input("Enter the Name of Favourite Movies: (comma seperatd) ")).split(",")
    
    movies_dic[name] = movies
```

```python
print(movies_dic)
```

```python
#2
all_movies = set()
for key,values in movies_dic.items():
    all_movies.update(values)
all_movies
```

```python
#3
movies_list = []
for key,values in movies_dic.items():
    movies_list.extend(values)
movies_list
```

```python
movies_freq = {}
for i in movies_list :
    if(i in movies_freq.keys()):
        movies_freq[i] += 1
    else:
        movies_freq[i] = 1
print(movies_freq)  
```
