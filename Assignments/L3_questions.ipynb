```python
#30
product_names = ("Laptop", "Mobile", "Tablet", "Headphones")
prices = (55000, 20000, 30000, 2000)
products = list(zip(product_names, prices))
```

```python

sorted_products = sorted(products, key=lambda x: x[1], reverse=True)

for product in sorted_products:
    print(product)
```
