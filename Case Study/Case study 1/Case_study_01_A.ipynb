```python
employee = {}
n = int(input("Enter the Number of employees :"))

for i in range(n):
    name = str(input("Enter the Name of Employee: "))
    no_of_skills = int(input("Enter the Number of skills :"))
    skills=[] ;
    for j in range(no_of_skills):
        skill = (str(input("Enter the skill of Employee: ")))
        skills.append(skill)

    employee[name] = skills
```

```python
for key, value in employee.items():
    if("Python" in value and "SQL" in value):
        print(key)
```

```python
all_skills = set()
for key, value in employee.items():
    all_skills.update(value)
```
