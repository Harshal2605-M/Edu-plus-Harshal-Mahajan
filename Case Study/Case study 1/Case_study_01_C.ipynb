```python
University = {}
n = int(input("Enter the Number of Students :"))

for i in range(n):
    courses = ()
    name = str(input("Enter the Name of Student: "))
    courses = str(input("Enter the Name of Courses: (comma seperatd) ")).split(",")
    
    University[name] = courses
```
