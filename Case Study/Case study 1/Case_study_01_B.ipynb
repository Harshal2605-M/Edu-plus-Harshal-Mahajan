```python
wh_1 = {}
wh_2 = {}

n_1 = int(input("Enter the Number of Products in wh_1 :"))
n_2 = int(input("Enter the Number of Products in wh_2 :"))

print("For Warhouse 1 :")
for i in range(n_1):
    product_id = int(input("Enter the Product Id: "))
    product_name = str(input("Enter the Name of product: "))
    product_quantity = int(input("Enter the quantity of product: "))
    wh_1[product_id] = product_name, product_quantity 

print("For Warhouse 2 :")
for i in range(n_2):
    product_id = int(input("Enter the Product Id: "))
    product_name = str(input("Enter the Name of product: "))
    product_quantity = int(input("Enter the quantity of product: "))   
    wh_2[product_id] = product_name, product_quantity 
```

```python
wh_1_products = set()
for key,values in wh_1.items():
    name = wh_1[key][0]
    wh_1_products.add(name)
```

```python
wh_1_products
```

```python
wh_2_products = set()
for key,values in wh_2.items():
    name = wh_2[key][0]
    wh_2_products.add(name)
```

```python
wh_2_products
```

```python
comman_products = wh_1_products.intersection(wh_2_products)
print("Comman Products :", comman_products)
```

```python
unique_products = wh_1_products.union(wh_2_products)
print("Unique Products :", unique_products)
```

```python
merged = wh_1.copy()

for product_id, (name, qty) in wh_2.items():
    if product_id in merged:
        old_name, old_qty = merged[product_id]
        merged[product_id] = (name, old_qty + qty)
    else:
        merged[product_id] = (name, qty)

print(merged)
```

```python
sorted_dic = dict(sorted(merged.items(), key = lambda a : a[1][1], reverse = True))
```
