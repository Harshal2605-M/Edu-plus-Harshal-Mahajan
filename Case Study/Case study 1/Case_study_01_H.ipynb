```python
data = {}

n = int(input("Enter No of Participants :"))

for i in range (n):
    sports = []
    name = str(input("Enter the name :"))
    sports = str(input("Enter the sports :")).split(",")
    data[name] = sports

print(data)
```

```python
#1
for key,value in data.items():
    if("Cricket" in value and "Football" in value):
        print(key)
```

```python
#2
unique = set()
for key, value in data.items():
    unique.update(value)
print(unique)
```

```python
#3
sports_list = []
for key,values in data.items():
    sports_list.extend(values)
sports_list
```

```python
sports_freq = {}
for i in sports_list:
    if(i in sports_freq.keys()):
        sports_freq[i] += 1
    else:
        sports_freq[i] = 1
print(sports_freq)  
```
