```python
data = {}

n1 = int(input("Enter number of travellers from Agent1: "))
n2 = int(input("Enter number of travellers from Agent2: "))

print("Data from Agent1 :")
for i in range(n1) :
    name = str(input("Enter the name of traveller :"))
    destination = str(input("Enter the name of destination:"))
    date = str(input("Enter the date (dd/mm/yyyy) :"))

    data[name] = destination, date


print("Data from Agent2 :")
for i in range(n2) :
    name = str(input("Enter the name of traveller :"))
    destination = str(input("Enter the name of destination:"))
    date = str(input("Enter the date (dd/mm/yyyy) :"))

    data[name] = destination, date

data
```

```python
#2
unique = set()
for key, value in data.items():
    dest = data[key][0]
    unique.add(dest)
unique
```

```python
#3
count_destination_frequency = {}
for key, value in data.items():
    dest = data[key][0]
    if(dest in count_destination_frequency.items()):
        count_destination_frequency[dest] += 1
    else:
        count_destination_frequency[dest] = 1
print(count_destination_frequency)
```

```python
count = 0
for key, value in count_destination_frequency.items():
    if(count < 3):
        print(key)
        count += 1
    else:
        break   
```
