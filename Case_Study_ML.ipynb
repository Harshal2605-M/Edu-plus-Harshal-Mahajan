```python
df['Rating'].fillna(df['Rating'].median(), inplace = True)
df
```

```python
df['Review'] = df['Review'].fillna("No review provided")
df
```

```python
# Q4: Create a new column 'TotalAmount' by multiplying Quantity and Price
# Add another column 'DiscountedPrice' that applies 10% discount on orders above $20
# Use conditional selection to filter orders where TotalAmount > $25

df['TotalAmount'] = df['Quantity']*df['Price']
df
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price'] = df['Price'].fillna(df['Price'].mean())
df
```

```python
df['TotalAmount'] = df['Quantity']*df['Price']
df
```

```python
df['DiscountedPrice'] = df['TotalAmount'] - df['TotalAmount']*0.1
df
```

```python
df_more_than_150 = df[df['TotalAmount']>25]
df_more_than_150
```
