```python
# 7 Create a nested function structure with three levels and print variables from local enclosing and global scopes
g_var = "global"

def outer():
    o_var = "outer"

    def middle():
        m_var = "middle"

        def inner():
            print("Local (inner):", m_var)
            print("Enclosing (outer):", o_var)
            print("Global:", g_var)

        inner()

    middle()

outer()
```

```python
# 13 Demonstrate how enclosing variable stores the running total for a counter function using nonlocal

def outer():
    a=8
    print("\nbefore counter the a is :",a)

    def counter(num):
        nonlocal a
        for i in range(1,num):
            a+=1
        print("\nAfter counter the a is :",a)

    counter(10)
outer()
```

```python
# 26 Create a nested function example where inner function modifies enclosing variable to accumulate running total of expenses
def expenses():
    totalexpenses=0
    def calaulate(*amount):
        nonlocal totalexpenses
        totalexpenses=sum(amount)
        return totalexpenses
    return calaulate
exp=expenses()
print("the total expenses are :",exp(1000,2000,3000))
```

```python
# 29 Write a function with both *args and **kwargs to simulate order details containing items and their prices

def order_details(*args, **kwargs):
    print("Order Summary")
    print("-" * 30)
    
    total = 0
    for item in args:
        price = kwargs.get(item, 0)
        print(f"{item:<15} : ${price:.2f}")
        total += price

    print("-" * 30)
    print(f"{'Total':<15} : ${total:.2f}")

order_details(
    "Burger", "Fries", "Cola",
    Burger=5.99,
    Fries=2.49,
    Cola=1.50
)
```
