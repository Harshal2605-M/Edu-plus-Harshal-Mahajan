```python
day1 = set(input("Enter students for Day 1 (space separated): ").split(","))
day2 = set(input("Enter students for Day 2 (space separated): ").split(","))
day3 = set(input("Enter students for Day 3 (space separated): ").split(","))
day4 = set(input("Enter students for Day 4 (space separated): ").split(","))
day5 = set(input("Enter students for Day 5 (space separated): ").split(","))
```

```python
# 1
present_every_day = day1 & day2 & day3 & day4 & day5
print(present_every_day)
```

```python
# 3
total_students = day1 | day2 | day3 | day4 | day5
print(total_students)
```

```python
# 2
absent_at_least_one_day = total_students - present_every_day
print(absent_at_least_one_day)
```
